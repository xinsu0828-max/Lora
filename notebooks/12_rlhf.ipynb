## 第 3 步: PPO (概念理解)

PPO 循环 (trl `PPOTrainer`):
1. policy 模型对 prompt 生成回答;
2. 奖励模型给回答打分;
3. PPO 用分数更新 policy, 并用 **KL 惩罚** 约束 policy 不偏离 SFT 太远 (防 reward hacking);
4. 重复。

PPO 流水线复杂、显存需求高、调参不稳定，免费 T4 难以稳定跑通，故本项目用 **DPO** 替代
(DPO 在数学上等价于 RM+PPO 的化简)。如需深入，可研究 trl 的 `PPOTrainer` / `GRPOTrainer`。
